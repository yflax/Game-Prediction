```python
# set X and y
select = [ 'elapsed_time',	'event_name',	 'room_coor_x',
       'room_coor_y', 'screen_coor_x', 'screen_coor_y' ,'fqid','fullscreen',	'hq',	'music']
X = merged_train2[select]
y = merged_train2['correct']
```

```python
X.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X,y,test_size=.2, random_state=2 )
```

```python
# create preprocessing pipeline
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
numericals = [ 'elapsed_time', 'room_coor_x','room_coor_y', 'screen_coor_x', 'screen_coor_y' ]

num_change = ColumnTransformer(transformers=[('Scale', StandardScaler(), numericals)])

pipe = Pipeline(steps=[('col_change', num_change),
        ('clf', RandomForestClassifier())])

pipe.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred_val = pipe.predict(X_val)
accuracy = accuracy_score(y_val,y_pred_val)
print(f"Accuracy: {accuracy:.2f}")
```
